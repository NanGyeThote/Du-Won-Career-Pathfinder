```python
import json
import os
```

```python
with open('career_kb.json', 'rt', encoding='utf-8') as f_out:
    list_file = json.load(f_out)
```

```python
chunks_file = {}
for items in list_file:
    career = items['career']
    texts = items['text']

    if career not in chunks_file:
        chunks_file[career] = texts
    else:
        chunks_file[career] += ' ' + texts
```

```python
chunks_file
```

```python
with open('chunks-data.txt', 'w', encoding='utf-8') as f_in:
    json.dump(chunks_file, f_in, indent=4, ensure_ascii=False)
```
