```python
from mistralai import Mistral
from dotenv import load_dotenv
```

```python
load_dotenv()
api_key = os.getenv('MISTRAL_API_KEY')
```

```python
# Build LLM Client with Mistral
client = Mistral(api_key=api_key)
model = 'mistral-small-2506'

# Mistral Client
def llm(prompt):
    res = client.chat.complete(
        model = model,
        messages = [
            {
                'role': 'user',
                'content': prompt
            }
        ]
    )
    answer = res.choices[0].message.content
    return answer
```

```python
prompt = input("You: ")
```

```python
ans = llm(prompt)
```

```python
# Importing Documents
with open('career_kb.json', 'rt', encoding='utf-8') as f_out:
    documents = json.load(f_out)
```

```python
from minsearch import Index
```

```python
index = Index(
    text_fields = ['text'],
    keyword_fields = ['career']
)
```

```python
# Indexing the documents
index.fit(documents)
```

```python
def search(query):
    boost = {"career": 3.0, "section": 0.5}
    results = index.search(
        query = query,
        filter_dict = {'career': 'Data Engineer'},
        boost_dict = boost,
        num_results = 5
    )
    return results
```

```python
query = 'How to become Data Engineer?'
search_results = search(query)
```

```python
search_results
```

```python
# def build_prompt
prompt_template = """
You are a career recommendation agent. Answer the QUESTION based on the CONTEXT with appropriate phrases.
If the answer is not in CONTEXT, please decline the request politely.

<QUESTION>
{question}
</QUESTION>

<CONTEXT>
{context}
</CONTEXT>
""".strip()

def build_prompt(query, search_results):
    context = ""
    for doc in search_results:
        context = context + f"career: {doc['career']}\nanswer: {doc['text']}\n\n"

    prompt = prompt_template.format(question=query, context=context).strip()
    return prompt
```

```python
prompt = build_prompt(query, search_results)
```

```python
ans = llm(prompt)
print(ans)
```

```python
# build RAG
def rag(query):
    search_results = search(query)
    prompt = build_prompt(query, search_results)
    results = llm(prompt)
    return results
```

```python
print('--- Start of The Program ---\n')
while True:
    try:
        prompt = input("You: ")
        if prompt != 'exit' and prompt != 'quit':
            answer = rag(prompt)
            print(f"Bot: {answer}\n")
            print("---\n")
        else:
            print("Bot: Good Bye!\n")
            print("--- End of the Program ---")
            break
    except Exception as e:
        print(f"An error occured: {e}")
```
